```python
# initial imports
import plotly.express as px
import pandas as pd
import panel as pn
from panel.interact import interact
from pathlib import Path
import hvplot.pandas

```

```python
# Set the file paths for the CSVs using the Path class from the pathlib library
coal_csv_path = Path("./data/Coal_02_11_22-02_14_19.csv")
copper_csv_path = Path("./data/Copper_02_14_22-02_14_19.csv")
iron_ore_csv_path = Path("./data/Iron Ore_02_14_22-07_13_20.csv")
lead_csv_path = Path("./data/Lead_02_14_22-02_14_19.csv")
gas_csv_path = Path("./data/Natural Gas (Henry Hub)_02_14_22-02_14_19.csv")
oil_csv_path = Path("./data/Oil (Brent)_02_11_22-02_14_19.csv")
palladium_csv_path = Path("./data/Palladium_02_14_22-02_14_19.csv")
platinum_csv_path = Path("./data/Platinum_02_14_22-02_14_19.csv")
uranium_csv_path = Path("./data/Uranium_02_10_22-07_14_20.csv")
```

```python
# Read the data, set the `date` as a datetime index
coal_df = pd.read_csv(coal_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
coal_df = coal_df.drop(columns=['Open', 'High', 'Low', 'Volume'])
#copper_df = pd.read_csv(copper_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)

iron_df = pd.read_csv(iron_ore_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
iron_df = iron_df.drop(columns=['Volume', 'Open', 'High', 'Low'])
lead_df = pd.read_csv(lead_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
lead_df = lead_df.drop(columns=['Volume', 'Open', 'High', 'Low'])

gas_df = pd.read_csv(gas_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
gas_df = gas_df.drop(columns=['Volume', 'Open', 'High', 'Low'])
oil_df = pd.read_csv(oil_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
oil_df = oil_df.drop(columns=['Volume', 'Open', 'High', 'Low'])
palladium_df = pd.read_csv(palladium_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
palladium_df = palladium_df.drop(columns=['Volume', 'Open', 'High', 'Low'])
platinum_df = pd.read_csv(platinum_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
platinum_df = platinum_df.drop(columns=['Volume', 'Open', 'High', 'Low'])
uranium_df = pd.read_csv(uranium_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
uranium_df = uranium_df.drop(columns=['Volume', 'Open', 'High', 'Low'])

# Combine the DataFrames by column and perform an inner join. 
combined_df = pd.concat([coal_df, iron_df, lead_df, gas_df, oil_df, palladium_df, platinum_df, uranium_df], axis="columns", join="inner")
combined_df

```

```python

# Sort datetime index in ascending order (past to present)
combined_df.sort_index(inplace=True)

# Drop extra columns and only leave the close columns and associated values
#combined_df.drop(columns=['Volume', 'Open', 'High', 'Low'], inplace=True)
combined_df
```

```python
# Rename the columns to match the corresponding stock
combined_df.columns = ['Coal', 'Iron Ore', 'Lead', 'Natural Gas', 'Oil', 'Palladium', 'Platinum', 'Uranium']
combined_df.head()
```

```python
# Calculate daily returns of closing prices for each column
daily_returns = combined_df.pct_change()
daily_returns.head()
```
